```python
import random as r
from math import sqrt

## estimating the values of the pi 
def distance(a,b):
    return sqrt(a**2 + b**2)

def estimating_pi(total_darts):
    inside_circle = 0
    
    for i in range(total_darts):
        x = r.random()
        y = r.random()
        
        if distance(x,y) < 1.0:
            inside_circle = 1
    pi = (inside_circle / total_darts)*4
    return pi

## random walk
def get_action():
    ## random movements on x and y cordinate 
    return random.choice([(0,1),(0,-1),(1,0),(-1,0)])

def random_walk(n):
    '''return a path after the n numbers of '''
    path =[]
    
    # initial values 
    x, y = 0, 0
    for i in range(n):
    
        (dx,dy) = get_action()

        x = x+dx
        y = y+dy

        path.append((x,y))
    
    return path
               
walk = random_walk(2000)
```

```python
num_of_simulation = int(10000000)
pi = estimating_pi(num_of_simulation)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 

plt.figure(num=None, figsize=(15,10))

point_color = list(range(2000))
x_numbers = [x[0] for x in walk]
y_numbers = [y[1] for y in walk]

plt.scatter(x_numbers,y_numbers, marker='.', s=20, c=point_color)

plt.show()
```
